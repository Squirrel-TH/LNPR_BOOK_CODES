```python
import pandas as pd
data = pd.read_csv("sensor_data_200.txt",delimiter=" ",header=None,names=("date","tiem","ir","lidar"))
data
```

```python
print(data["lidar"][0:5])
```

```python
import matplotlib.pyplot  as plt
data["lidar"].hist(bins=max(data["lidar"])-min(data["lidar"]),align='left')
plt.show()
```

```python
mean1 = sum(data["lidar"].values)/len(data["lidar"].values)
mean2 = data["lidar"].mean()
print(mean1, mean2)
```

```python
data["lidar"].hist(bins = max(data["lidar"]) - min(data["lidar"]), color="orange",align='left')
plt.vlines(mean1,ymin=0,ymax=5000,color="red")
plt.show()
```

```python
# å®šç¾©ã‹ã‚‰è¨ˆç®—
zs = data["lidar"].values
mean = sum(zs)/len(zs)
diff_square = [(z-mean)**2 for z in zs]

sampling_var = sum(diff_square)/(len(zs)) # æ¨™æœ¬åˆ†æ•£
unbiasd_var = sum(diff_square)/(len(zs)-1) # ä¸ååˆ†æ•£

print(sampling_var)
print(unbiasd_var)
```

```python
# pandasã‚’ä½¿ç”¨
pandas_sampling_var = data["lidar"].var(ddof=0) # æ¨™æœ¬åˆ†æ•£
pandas_default_var = data["lidar"].var() # ä¸ååˆ†æ•£

print(pandas_sampling_var)
print(pandas_default_var)
```

```python
# Numpyã‚’ä½¿ç”¨
import numpy as np

numpy_default_var = np.var(data["lidar"]) # æ¨™æœ¬åˆ†æ•£
numpy_nubiased_var = np.var(data["lidar"],ddof=1) # ä¸ååˆ†æ•£

print(numpy_default_var)
print(numpy_nubiased_var)
```

```python
import math

# å®šç¾©ã‹ã‚‰è¨ˆç®—
stddev1 = math.sqrt(sampling_var)
stddev2 = math.sqrt(unbiasd_var)

# Pandasã‚’ä½¿ç”¨
pandas_stddev = data["lidar"].std()

print(stddev1)
print(stddev2)
print(pandas_stddev)
```

```python
freqs = pd.DataFrame(data["lidar"].value_counts())
freqs.transpose()
```

```python
freqs["probs"] = freqs["lidar"]/len(data["lidar"])
freqs.transpose()
```

```python
sum(freqs["probs"])
```

```python
freqs["probs"].sort_index().plot.bar()
plt.show()
```

```python
def drawing():
    return freqs.sample(n=1, weights="probs").index[0]

drawing()
```

```python
samples = [drawing() for i in range(10000)]
simulated = pd.DataFrame(samples, columns=["lidar"])
p = simulated["lidar"]
p.hist(bins = max(p) - min(p), color="orange", align='left')
plt.show()
```
